```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
import math

# 1) Initially assigning value betweem 60 to 70
data = [i for i in range(60,70)]
#randemly chosing 1000 values
[data.append(random.choice(data)) for i in range(1000)]
print(data)
```

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
import math

# making pandas series
df = pd.Series(data)
df
```

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
import math

# 2) Calculating mean
mean_val = df.mean()
print(f'Mean : {mean_val}')
```

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
import math

# Calculating Standard deviation
std_val = df.std()
print(f'Standard deviation : {std_val}')
```

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
import math

# 3) Ploting Histogram
frq = collections.Counter(data)
plt.hist(data)

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Values between 60 and 70')
plt.show()
```

```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
import math

sample = []
smean = []
# 4) & 5)chosing random value and selecting one by one, and calculating mean for each size of list
for i in range(30):
    for j in range(30):
        val = random.choice(data)
        sample.append(val)
    temp_mean = np.mean(sample)
    smean.append(temp_mean)
print(smean)
```
